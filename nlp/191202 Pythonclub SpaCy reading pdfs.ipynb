```python
import spacy 
nlp = spacy.load("en_core_web_sm")
f = open('C:/Users/oskar/Downloads/Report Liability for AI and other digital technologies.pdf.txt', encoding = 'utf-8')
text = f.read()
doc = nlp(text)
```

```python
print('Given text contains the following entities')
for ent in doc.ents:
    if ent.label_=='LAW': #try changing 'LAW' to 'ORG', or remove the if statement to see all entities
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from spacy import displacy
displacy.serve(doc, style="ent")
```
